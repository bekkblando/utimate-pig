```python
import matplotlib.pyplot as plt
import player
from player import Game, Player, Strategy_Player, Random_Player
import statistics as stats
```

```python
def str0start():
    gm1 = Game()
    player1 = Player(1)
    for number in range(7):
        gm1.turn(player1)
    return player1.player_points

def str1start():
    gm1 = Game()
    player1 = Strategy_Player(100)
    for number in range(7):
        gm1.turn(player1)
    return player1.player_points

def str2start():
    gm1 = Game()
    player1 = Random_Player()
    for number in range(7):
        gm1.turn(player1)
    return player1.player_points
    
data = [str2start() for item in range(2000)]
# print(data)
```
