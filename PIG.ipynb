```python
def str0start():
    gm1 = Game()
    player1 = Player(1)
    for number in range(7):
        gm1.turn(player1)
    return player1.player_points

def str1start():
    gm1 = Game()
    player1 = Strategy_Player(10)
    for number in range(7):
        gm1.turn(player1)
    return player1.player_points

def str2start():
    gm1 = Game()
    player1 = Random_Player()
    for number in range(7):
        gm1.turn(player1)
    return player1.player_points
    
data = [str1start() for item in range(2000)]
print(data)
```
